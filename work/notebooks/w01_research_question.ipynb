**Decision:** Which decaying or stale pages should an SEO content editor review and refresh first?

**Who acts and what they do:** SEO editors and content strategists use this ranked queue to allocate their limited weekly editing and updating hours.

**Cost of a wrong call**: A false positive (flagging a stable page) wastes valuable editor time. A false negative (missing a declining page) leads to uncaptured traffic decay that could have been mitigated.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

url = "https://raw.githubusercontent.com/huzaifa505694/FLY-RANK-INTERN/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

total_pages = df.shape[0]
declining_rate = df["trend_direction"].str.lower().eq("down").mean()
median_age = df["content_age_days"].median()

print(f"Total rows in dataset: {total_pages}")
print(f"Observed declining rate ('down'): {declining_rate:.3f}")
print(f"Observed median content age (days): {median_age:.0f}")
```
