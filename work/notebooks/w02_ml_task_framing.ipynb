## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

Our provisional ML task type for this assignment is **Ranking and Scoring**. We are specifically developing an algorithmic priority score—a `refresh_priority_score`—designed to dynamically order and rank content items.

We chose this task type because the primary operational challenge for content strategists is not a binary yes/no question, but an abundance of content inventory where resources are scarce, meaning they need a sorted, prioritized queue of pages to address sequentially. Ranking aligns directly with real editorial workflows by ensuring that the most critical or high-opportunity decaying pages surface at the very top of the actionable list every single week.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

We are predicting a continuous composite opportunity/priority score that acts as a robust proxy for future traffic recovery potential. Rather than utilizing a circular or hand-written rule definition as our training target—which would only teach the model to memorize a human's arbitrary threshold—our target construction relies on multi-signal observed historical metrics.

Specifically, we integrate observed downward percentage trends (`trend_pct`), aging content duration (`content_age_days`), and baseline traffic exposure metrics from trailing windows. This ensures the model learns true behavioral patterns of content degradation and audience engagement rather than fitting to a pre-defined administrative label.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

To objectively evaluate and defend our ranking system, our primary success metric is **Precision@K** (evaluated specifically at $K = 20$ and $K = 50$). Precision@K measures the proportion of top-$K$ recommended content items that correspond to genuinely declining or high-opportunity performance metrics in subsequent validation slices.

A quantitative benchmark for a "good" operational model in this workspace is achieving a **Precision@20 score greater than 0.70**, indicating that at least 70% of the top priority items flagged by our ranking algorithm truly represent high-decay or high-value refresh candidates when cross-referenced with empirical performance outcomes.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A rigid, hand-written `if-statement` or static threshold rule fails completely in this domain because content performance signals interact in highly non-linear, multifaceted, and noisy ways across different clients and content categories. Factors like content age, average GSC position, traffic volume fluctuations, and engagement percentages do not follow clean, uniform cutoff boundaries.

An algorithmic ML or scoring approach successfully captures the subtle trade-offs and varying baseline contexts between different publishing environments, weighing multiple continuous signals simultaneously and smoothly rather than enforcing brittle, arbitrary boundaries that either overwhelm editors with false positives or miss critical decay patterns.
